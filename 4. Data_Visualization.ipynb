### Plotting: Area plots, Histograms, Bar charts, Pie Charts and Scatter Charts

A line graph is a basic plot that displays the relationship between two variables on each axis by connecting datapoints together with straight lines.

 To show magnitude, the segment between the line and the x-axis is filled which results in an area graph. 
 
 Histograms and bar charts are completely different plots that can be mistaken as similar. The former are charts used to represent the distribution of a group and use adjacent rectangular bars to display the frequency of intervals while the latter are charts that represent categories using equally spaced rectangular bars.

```python
#import plotting library
import matplotlib.pyplot as plt

plt.figure(figsize = (7,4))
```

```python
plt.xticks(rotation = 90)
```

```python
import pandas as pd
fuel_unit = pd.DataFrame({'unit':['BBL', 'GAL', 'GRAMSU', 'KGU', 'MCF', 'MMBTU', 'MWDTH', 'MWHTH', 'TON'],
                         'count':[7998, 84, 464, 110, 11354, 180, 95, 100, 8958]})
print(fuel_unit)
```

```python
import seaborn as sns
sns.barplot(data = fuel_unit, x = 'unit', y = 'count')
plt.xlabel('Fuel Unit')
```

Because of the extreme range of the values for the fuel unit, we can plot the barchart by taking the logarithm of the y-axis as follow

```python
g = sns.barplot(data = fuel_unit, x = 'unit', y = 'count')
g.set_yscale('log')
g.set_ylim(1, 12000)
plt.xlabel('Fuel Unit')
```
